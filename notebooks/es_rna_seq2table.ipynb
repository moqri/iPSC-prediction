```python
data='../data/'
esc='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE73nnn/GSE73211/suppl/GSE73211_nonisogenic.TPM.txt.gz'
esc2='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE73nnn/GSE73211/suppl/GSE73211_isogenic.TPM.txt.gz'
```

```python
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
df=pd.read_csv(esc,sep='\t',header=None,skiprows=1,index_col=0)
columns=pd.read_csv(esc,nrows=0,sep=' ').columns
columns=columns.str.replace("ID\t",'')
df.columns=columns
```

```python
df=df[[c for c in df.columns if 'iPS' not in c]]
df.shape
```

```python
df2=pd.read_csv(esc2,sep='\t',header=None,skiprows=1,index_col=0)
columns=pd.read_table(esc2,nrows=0).columns
columns=columns.str.replace("ID\t",'')
df2.columns=columns
df2=df2[[c for c in df2.columns if 'hESC' in c and 'Fib' not in c ]]
```

```python
df=df.merge(df2,left_index=True,right_index=True)
df.head()
```

```python
df.astype(float).round(2).T.to_csv(data+'es_tpm.csv.gz', compression='gzip')
```
